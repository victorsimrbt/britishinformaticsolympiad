```python
import math

interest,repayment = [float(val)/100 for val in input().split()]

repaid = 0.0
debt = 100.0
def threedp(n):
    if type(n) == int:
        return n
    integer,decimal = str(n).split('.')
    return float(integer+'.'+decimal[:2])

while debt > 0:
    debt *= (1+interest)
    debt = round(debt,5)
    debt = math.ceil(round(threedp(debt)*100))/100
    repay = max(repayment * debt,50.0)
    repay = min(repay,debt)
    repay = round(repay,5)
    repay = math.ceil(round(threedp(repay)*100))/100
    print(debt,repay,repaid)
    debt -= repay
    debt = round(debt,5)
    repaid += repay
    repaid = round(repaid,5)
    
repaid
```
